```python
import math
import numpy as np
```

```python
sin=np.array([math.sin(x) for x in range(200) ])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(sin[:50])
```

```python
X=[]
Y=[]
seq_len=50
num_records=len(sin)-seq_len #=200-50=150
for i in range(num_records-50):
    X.append(sin[i:i+50])
    Y.append(sin[i+50])
X=np.array(X)
Y=np.array(Y)
X=np.expand_dims(X,axis=2)
Y=np.expand_dims(Y,axis=1)
```

```python
X.shape,Y.shape
```

```python
learning_rate = 0.0001    
nepoch = 25               
T = 50                   # length of sequence
hidden_dim = 100         
output_dim = 1

bptt_truncate = 5
min_clip_value = -10
max_clip_value = 10
```

```python
U = np.random.uniform(0, 1, (hidden_dim, T))
W = np.random.uniform(0, 1, (hidden_dim, hidden_dim))
V = np.random.uniform(0, 1, (output_dim, hidden_dim))
print(U.shape,V.shape,W.shape)
```

```python
tup=(50,1)
test=np.zeros(tup)
test.shape
```

```python
def sigmoid(x):
    return 1/(1+(np.exp(-1*x)))
```

```python
for epoch in range(nepoch):
    loss=0
    for i in range(Y.shape[0]):
        x,y=X[i],Y[i]
        prev_s=np.zeros((hidden_dim,1))
        for t in range(T):
            new_input=np.zeros(x.shape)
            new_input[t]=x[t]
            matu=np.dot(U,new_input)
            matw=np.dot(W,prev_s)
            add=matu+matw
            s=sigmoid(add)
            mulv=np.dot(V,s)
            prev_s=s
        #error
        print("value going in loss="+str((y-mulv)**2/2))
        loss+=(y-mulv)**2/2
    print("loss at each epoch is "+str(loss))
        
   
```

```python
for epoch in range(nepoch):
    # check loss on train
    loss = 0.0
    
    # do a forward pass to get prediction
    for i in range(Y.shape[0]):
        x, y = X[i], Y[i]                    # get input, output values of each record
        prev_s = np.zeros((hidden_dim, 1))   # here, prev-s is the value of the previous activation of hidden layer; which is initialized as all zeroes
        for t in range(T):
            new_input = np.zeros(x.shape)    # we then do a forward pass for every timestep in the sequence
            new_input[t] = x[t]              # for this, we define a single input for that timestep
            mulu = np.dot(U, new_input)
            mulw = np.dot(W, prev_s)
            add = mulw + mulu
            s = sigmoid(add)
            mulv = np.dot(V, s)
            prev_s = s

    # calculate error 
        loss_per_record = (y - mulv)**2 / 2
        loss += loss_per_record
        print("losses at each epochs are:"+str(loss))
    loss = loss / float(y.shape[0])
```
